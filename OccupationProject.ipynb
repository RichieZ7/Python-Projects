```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

missing_values = ["Na"]
gender_employment_data = pd.read_csv("inc_occ_gender.csv", na_values=missing_values)
median_all = gender_employment_data["All_weekly"].median()
median_male = gender_employment_data["M_weekly"].median()
median_female = gender_employment_data["F_weekly"].median()
gender_employment_data["All_weekly"].fillna(median_all, inplace=True)
gender_employment_data["M_weekly"].fillna(median_male, inplace=True)
gender_employment_data["F_weekly"].fillna(median_female, inplace=True)
# gender_employment_data

X = gender_employment_data.drop(columns = ["Occupation"])
Y = gender_employment_data["Occupation"]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)

model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

score = accuracy_score(Y_test, predictions)
score
```
